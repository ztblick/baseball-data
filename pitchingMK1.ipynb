```python
import pandas as pd
import requests
import numpy as np
import statsapi
```

```python
url = 'http://statsapi.mlb.com/api/v1/schedule/games/?sportId=1'
r = requests.get(url=url).json()
dates = r['dates'][0]['games']
df = pd.DataFrame(dates)
df
```

```python
HOST = 'https://statsapi.mlb.com'
url = HOST + df.loc[0,'link']
r = requests.get(url=url).json()
player_df = pd.DataFrame(r['gameData']['players'])
player_df_all = pd.DataFrame(['players'])
player_df = player_df.T
player_df.T
```

```python
def get_hitsPer9(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "pitching", type="season", sportId=1)
    if player['position'] != 'P':
        return 0
    return float(player['stats'][0]['stats']['hitsPer9Inn'])

def get_inningsPitched(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "pitching", type="season", sportId=1)
    if player['position'] != 'P':
        return 0
    return float(player['stats'][0]['stats']['inningsPitched'])

def get_gamesStarted(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "pitching", type="season", sportId=1)
    if player['position'] != 'P':
        return 0
    return float(player['stats'][0]['stats']['gamesStarted'])

player_hitsPer9 = player_df.id.apply(get_hitsPer9)
player_df = player_df.assign(hitsPer9=player_hitsPer9.values)

player_inningsPitched = player_df.id.apply(get_inningsPitched)
player_df = player_df.assign(inningsPitched=player_inningsPitched.values)

player_gamesStarted = player_df.id.apply(get_gamesStarted)
player_df = player_df.assign(gamesStarted=player_gamesStarted.values)
```

```python
pitchers = player_df['hitsPer9'] != 0
pitcher_df = player_df.loc[pitchers]

prevPitchingData = pd.read_csv("22-23 Pitching Data.csv")


def get_last3HP9(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "pitching", type="season", sportId=1)
    num_seasons = 1
    penult_ult_hp9 = 0
    ip = 0
    pitch_df = prevPitchingData.loc[prevPitchingData['player_id'] == player_id]
    if not pitch_df.empty:
        for index, row in pitch_df.iterrows():
            penult_ult_hp9 += float(pitch_df['hit'].iloc[0])
            ip += float(pitch_df['p_formatted_ip'].iloc[0])
            num_seasons += 1
    if ip == 0:
        penult_ult_hp9 = 0
    else:
        penult_ult_hp9 = penult_ult_hp9 / ip * 9
    current = float(player['stats'][0]['stats']['hitsPer9Inn'])
    return (penult_ult_hp9 + current) / num_seasons

L3HP9 = pitcher_df.id.apply(get_last3HP9)
pitcher_df = pitcher_df.assign(L3SeasonHP9=L3HP9.values)

startingPitchers = pitcher_df['gamesStarted'] != 0
startingPitcher_df = pitcher_df.loc[startingPitchers]
startingPitcher_df.sort_values(by = ['hitsPer9'], ascending = True)

startingPitcher_df.T
```
