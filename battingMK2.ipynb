```python
import pandas as pd
import requests
import numpy as np
import statsapi
```

```python
url = 'http://statsapi.mlb.com/api/v1/schedule/games/?sportId=1'
r = requests.get(url=url).json()
dates = r['dates'][0]['games']
df = pd.DataFrame(dates)
df
```

```python
HOST = 'https://statsapi.mlb.com'
url = HOST + df.loc[0,'link']
r = requests.get(url=url).json()
player_df = pd.DataFrame(r['gameData']['players'])
player_df_all = pd.DataFrame(['players'])
player_df = player_df.T
player_df.T
```

```python
def get_avg(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "hitting", type="season", sportId=1)
    if player['position'] == 'P':
        return 0
    return float(player['stats'][0]['stats']['avg'])

def get_atBats(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "hitting", type="season", sportId=1)
    if player['position'] == 'P':
        return 0
    return float(player['stats'][0]['stats']['atBats'])


averages = player_df.id.apply(get_avg)
player_df = player_df.assign(avg=averages.values)

player_atBats = player_df.id.apply(get_atBats)
player_df = player_df.assign(atBats=player_atBats.values)
```

```python
# The cell below will get the batting averages and at bats of the players from seasons 22-24
```

```python
prevBattingData = pd.read_csv("22-23 Batting Data.csv")

def get_last3avg(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "batting", type="season", sportId=1)
    num_seasons = 2
    penult_ult_avg = 0
    bat_df = prevBattingData.loc[prevBattingData['player_id'] == player_id]
    if not bat_df.empty:
        for index, row in bat_df.iterrows():
            penult_ult_avg += float(bat_df['batting_avg'].iloc[0])
            num_seasons += 1
    current = float(player['stats'][0]['stats']['avg'])
    # current season avg multiplied by 2 in order to give greater weight to more recent performance
    return (penult_ult_avg + current * 2) / num_seasons

def get_last3AB(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "batting", type="season", sportId=1)
    num_seasons = 2
    penult_ult_atBats = 0
    bat_df = prevBattingData.loc[prevBattingData['player_id'] == player_id]
    if not bat_df.empty:
        for index, row in bat_df.iterrows():
            penult_ult_atBats += float(bat_df['ab'].iloc[0])
            num_seasons += 1
    current = float(player['stats'][0]['stats']['atBats'])
    return (penult_ult_atBats + current * 2) / num_seasons

L3AVG = player_df.id.apply(get_last3avg)
player_df = player_df.assign(L3SeasonAVG=L3AVG.values)

L3AB = player_df.id.apply(get_last3AB)
player_df = player_df.assign(L3SeasonAB=L3AB.values)
```

```python
totalBatScore = player_df.avg * player_df.atBats / 30
player_df = player_df.assign(totalBatScore=totalBatScore.values)
player_df.T
```
