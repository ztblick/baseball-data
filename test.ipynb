### Testing out the MLB Stats API

```python
# Import necessary modules
import pandas as pd
import requests
import numpy as np
```

How to call the API for the games on each day:

```python
url = 'http://statsapi.mlb.com/api/v1/schedule/games/?sportId=1'
r = requests.get(url=url).json()
dates = r['dates'][0]['games']
df = pd.DataFrame(dates)
df
```

```python
df.loc[6,'teams']
```

```python
# Getting the players from today's first game:
HOST = 'https://statsapi.mlb.com'
url = HOST + df.loc[0,'link']
r = requests.get(url=url).json()
player_df = pd.DataFrame(r['gameData']['players'])
player_df.T
```

```python
# player_df = player_df.T
player_df.columns
foreign_players = player_df['birthCountry'] != 'USA'
player_df.loc[foreign_players]
```

Trying out the python MLB_Stats_API library:

```python
import statsapi
```

```python
print('The A\'s won %s games in 2018.' % sum(1 for x in statsapi.schedule(team=133,start_date='03/29/2018',end_date='12/31/2018') if x.get('winning_team','')=='Oakland Athletics'))
```

```python
most_recent_game_id = statsapi.last_game(133)
a = statsapi.boxscore(most_recent_game_id)
b = statsapi.linescore(most_recent_game_id)
# print(a)
print(b)
```

```python
# Get standings as a string to be printed
standings = statsapi.standings(leagueId="103,104", division="all", include_wildcard=True, season=None, standingsTypes=None, date=None)
print(standings)
```

```python
# Get standings as a dictionary:
standings_data = statsapi.standings_data(leagueId="103,104", division="all", include_wildcard=True, season=None, standingsTypes=None, date=None)
standings_data
```

```python
# Get team info:
teams = statsapi.lookup_team("", activeStatus="Y")
teams = pd.DataFrame(teams)
teams.sort_values(by=['id'])
```

```python
# Write out the teams data to a csv file
teams.to_csv(path_or_buf="baseball-data/team_data.csv", index=False)
```
