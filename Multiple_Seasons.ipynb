## This notebook is to demonstrate how to retrieve game stats for specific teams over a given period of time

```python
import pandas as pd
import requests
import numpy as np
import statsapi
from datetime import datetime, timedelta
TODAY = datetime.today().strftime('%m/%d/%Y')
LAST_YEAR = (datetime.today() - timedelta(days=365)).strftime('%m/%d/%Y')
```

```python
# Gets one player's stats for one game
data = statsapi.get('person_stats', {'personId' : '665742', 'gamePk' : '745717'})
```

```python
# Gets games in a certain range
games = statsapi.schedule(start_date=LAST_YEAR,end_date=TODAY,team=143,opponent=121)
```

```python
# In retrospect, this is not the best way to do this. It was much easier to gather
# data from previous seasons using Savant: https://baseballsavant.mlb.com/leaderboard
```

```python
for game in games:
    game_data = statsapi.boxscore_data(game['game_id'], timecode=None)
    players = pd.DataFrame(game_data['away']['players'])
    # print(game_data['ID'+str(game['game_id'])])
# for player_id in list(players.keys()):
#     print(players[player_id]['seasonStats'])
players = players.T
```
