```python
def get_hitsPer9(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "pitching", type="season", sportId=1)
    if player['position'] != 'P':
        return 0
    return float(player['stats'][0]['stats']['hitsPer9Inn'])

def get_inningsPitched(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "pitching", type="season", sportId=1)
    if player['position'] != 'P':
        return 0
    return float(player['stats'][0]['stats']['inningsPitched'])

def get_gamesStarted(player_id):
    player = statsapi.player_stat_data(int(player_id), group= "pitching", type="season", sportId=1)
    if player['position'] != 'P':
        return 0
    return float(player['stats'][0]['stats']['gamesStarted'])

player_hitsPer9 = player_df.id.apply(get_hitsPer9)
player_df = player_df.assign(hitsPer9=player_hitsPer9.values)

player_inningsPitched = player_df.id.apply(get_inningsPitched)
player_df = player_df.assign(inningsPitched=player_inningsPitched.values)

player_gamesStarted = player_df.id.apply(get_gamesStarted)
player_df = player_df.assign(gamesStarted=player_gamesStarted.values)
```

```python
pitchers = player_df['hitsPer9'] != 0
pitcher_df = player_df.loc[pitchers]

startingPitchers = player_df['gamesStarted'] != 0
startingPitcher_df = player_df.loc[startingPitchers]
startingPitcher_df.sort_values(by = ['hitsPer9'], ascending = True)
```
